## 3. Pie chart

is a circle that is divided into sections or wedges according to the
percentage of frequencies in each category of the distribution.


````python
df = pd.read_csv('Dataset/Bank Customer Churn Prediction.csv')
````

````python
df
````

````python
df.value_counts()
````

````python
df[['country']].value_counts()
````

````python
labels=df['country'].value_counts().index
````

````python
fig, ax = plt.subplots(figsize = (6, 6), dpi = 100)
ax.pie(x = df['country'].value_counts(), 
       labels=df['country'].value_counts().index,
      autopct='%1.2f%%',
      explode= [0, 0.1, 0],
      startangle=270)
ax.legend(df['country'].value_counts().index)
plt.savefig('Pie_chart.png', dpi = 1000)
plt.show()
````

````python
fig, ax = plt.subplots( figsize = (6, 6), dpi = 100)
ax.pie(x = df['country'].value_counts(),
       labels = df['country'].value_counts().index,
      autopct='%1.1f%%',
      explode= [0.1, 0, 00],
      startangle=90)
ax.set_title('Distribution of Country')
ax.legend(df['country'].value_counts().index, loc='upper left')
fig.savefig('pie.png', dpi = 1000)
plt.show()
````

## 4. Donut Plot
<a id='donut_plot'></a>


````python
fig, ax = plt.subplots(figsize = (6, 6), dpi = 100)
# Pie Chart
ax.pie(df['country'].value_counts(), labels=df['country'].value_counts().index,
        autopct='%1.1f%%', pctdistance=0.85)
 
# draw circle
centre_circle = plt.Circle((0, 0), 0.7, fc='white')
fig = plt.gcf()
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
fig.savefig('donutplot.png', dpi = 1000)
plt.show()
````

## 5. Histrogram

**Histrogram** is a graph that shows the frequency of numerical data using rectangles. The height of a rectangle (the vertical axis) represents the distribution frequency of a variable (the amount, or how often that variable appears).

**Histogram** is a graph that displays the data by using contiguous vertical bars
(unless the frequency of a class is 0) of various heights to represent the
frequencies of the classes.


````python
df
````

````python
df.isnull().sum()
````

````python
df.head()
````

````python
df['churn'].replace({1:'Yes', 0:'No'}, inplace=True)
````

````python
plt.hist()
````

````python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize = (16, 6))
sns.histplot(df['age'], ax = ax[0])
sns.histplot(df['age'], ax = ax[1])
sns.histplot(df['estimated_salary'], ax = ax[2])

ax[0].set_title('A')
ax[0].set_xticks(np.arange(15, 101, 5))
ax[0].set_yticks(np.arange(0, 901, 50))

ax[1].set_title('B')

ax[2].set_title('C')

plt.show()
````

````python
bins = 40
fig, ax = plt.subplots(1, 2, figsize = (16, 8), dpi = 100)
sns.histplot(df['age'], ax = ax[0], label = 'Distribution of Age', bins = bins)
sns.histplot(df[df['churn']=='Yes']['age'], ax = ax[0], color = 'r', 
             label = 'Distribution of age with Exited Customer', bins= bins)

ax[0].set_title('Age Distribution for Exited and Non-Exited Customers', fontsize=14)
ax[0].set_ylabel('Number of Customers', fontsize=14)
ax[0].set_xticks(np.arange(18,90,5))
ax[0].set_yticks(np.arange(0,1100,100))

ax[1].set_title('Age Distribution for Exited and Non-Exited Customers', fontsize=14)
sns.histplot(df['age'], ax = ax[1], label = 'Distribution of Age', bins = bins)
sns.histplot(df[df['churn']=='No']['age'], ax = ax[1],
             label = 'Distribution of age with Non-Exited Customers', bins = bins)

ax[1].set_ylabel('Number of Customers', fontsize=14)
ax[1].set_xticks(np.arange(18,90,5))
ax[1].set_yticks(np.arange(0,1100,100))

for i in ax:
    i.legend()
    i.spines[['top', 'right']].set_visible(False)
    i.grid(False)
plt.tight_layout()
fig.savefig('Photos/Distribution_of_age.png', dpi = 2000)
plt.show()
````

## 6. Measures of position: quartiles
### Box Plot
<a id='box_plot'></a>
**Box plots** provide a quick visual summary of the variability of values in a dataset. They show the median, upper and lower quartiles, minimum and maximum values, and any outliers in the dataset. Outliers can reveal mistakes or unusual occurrences in data.

````python
fig, ax = plt.subplots(figsize = (4, 10))
sns.boxplot(df['estimated_salary'])
ax.set_yticks(np.arange(10000, 2.1*(10**5), 10000))
ax.grid(False)
````

## Violin Plot
<a id='Violin_plots'></a>

### How to Interpret Violin Plots?

Similar to `histograms`, `violin plots` show the shape of a data set. However, unlike histograms, violin plots show the shape of a data set by using a Probability Density Function (PDF), or a density plot, which is effectively a smoothed-over histogram. 


<img src="Photos/violin-plot-describe.png" alt="Control Structure" width="50%" style="display: block; margin: 0 auto;">

````python
fig, ax = plt.subplots(figsize = (4, 6))
sns.violinplot(df['estimated_salary'])
ax.set_yticks(np.arange(10000, 2.1*(10**5), 10000))
ax.grid(False)
````
